```python
from sklearn.cluster import KMeans
```

```python
X = df[['Murder']]

# Apply K-means clustering with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)

# Assign cluster labels to the original dataframe
df['Cluster'] = kmeans.labels_

# Define cluster names
cluster_names = {0: 'Sensitive Areas', 1: 'Moderate Areas', 2: 'Peaceful Areas'}

# To ensure the cluster names are assigned correctly, we need to map them based on the cluster's mean values.
# Get mean murders for each cluster to assign appropriate names
cluster_means = df.groupby('Cluster')['Murder'].mean().sort_values()

# Assign human-readable names to clusters based on their mean values
ordered_clusters = cluster_means.index.tolist()
cluster_map = {ordered_clusters[0]: 'Peaceful Areas', 
               ordered_clusters[1]: 'Moderate Areas', 
               ordered_clusters[2]: 'Sensitive Areas'}

df['Cluster Name'] = df['Cluster'].map(cluster_map)

# Print the final dataframe
print(df[['Murder','States/UTs','District','Total Cognizable IPC crimes','Cluster Name']])

# Optionally visualize the clusters
plt.scatter(df['District'], df['Murder'], c=df['Cluster'])
plt.xlabel('District')
plt.ylabel('Murder')
plt.title('Districts Clustered by Number of Murder')
plt.show()
```
